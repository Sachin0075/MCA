```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Read the image in grayscale
f = cv2.imread('cubic.jpg', 0)
plt.imshow(f, cmap="gray")

plt.title('f(x,y)')
plt.axis('off')
plt.show()

# Apply Fourier Transform
F = np.fft.fft2(f)

plt.imshow(np.log1p(np.abs(F)), cmap="gray")
plt.title('F')
plt.axis('off')
plt.show()

# Shift the zero-frequency component to the center
Fshift = np.fft.fftshift(F)

plt.imshow(np.log1p(np.abs(Fshift)), cmap='gray')
plt.title('F(U,V)')
plt.axis('off')
plt.show()

# Create a mask for Ideal Low Pass Filter (ILPF)
M, N = f.shape
H = np.zeros((M, N), dtype=np.float32)
D0 = 50  # Cutoff frequency

for u in range(M):
    for v in range(N):
        D = np.sqrt((u - M/2)**2 + (v - N/2)**2)
        if D <= D0:
            H[u, v] = 1
        else:
            H[u, v] = 0

plt.imshow(H, cmap='gray')
plt.title('H')
plt.axis('off')
plt.show()

# Apply the filter in the frequency domain
Gshift = Fshift * H

plt.imshow(np.log1p(np.abs(Gshift)), cmap="gray")
plt.title('G(U,V) = F(U,V) * H(U,V)')
plt.axis('off')
plt.show()

# Inverse shift and inverse Fourier Transform
G = np.fft.ifftshift(Gshift)

plt.imshow(np.log1p(np.abs(G)), cmap='gray')
plt.title('Ginverse(U,V)')
plt.axis('off')
plt.show()

# Get the inverse Fourier Transform to obtain the smoothed image
g = np.abs(np.fft.ifft2(G))

plt.imshow(g, cmap='gray')
plt.title('g(x,y) = Image Smoothing ILPF')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('Me.jpeg')
new_img = cv2.resize(img, (500, 500))
flip_img=np.flipud(new_img)

flip_img_rgb = cv2.cvtColor(flip_img, cv2.COLOR_BGR2RGB)

plt.imshow(flip_img_rgb)
plt.title('Resized Image')
plt.axis('off')
plt.show()

# cv2.imshow('Resized Image', flip_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image= cv2.imread('Me.jpeg')
img=cv2.resize(image,(500,500))
rows,cols,channel=img.shape
result=img.copy()
for i in range(rows):
    k=0
    for j in range(cols-1,-1,-1):
        temp=img[i,k].copy()
        result[i,k]=img[i,j]
        result[i,j]=temp
        k+=1
cv2.imshow('Mirror Image',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
